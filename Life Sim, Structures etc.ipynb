```python
import matplotlib.pyplot as plt
import numpy as np

# Function to draw a circle (representing the viral envelope)
def draw_circle(ax, center, radius, color='lightblue', edge_color='blue', alpha=0.5):
    circle = plt.Circle(center, radius, color=color, edgecolor=edge_color, alpha=alpha)
    ax.add_patch(circle)

# Function to draw the capsid as a polygon (simple icosahedral representation)
def draw_capsid(ax, center, radius, color='lightgreen', edge_color='green', alpha=0.6):
    num_sides = 20  # Icosahedral shape
    angles = np.linspace(0, 2 * np.pi, num_sides, endpoint=False)
    vertices = [(center[0] + radius * np.cos(angle), center[1] + radius * np.sin(angle)) for angle in angles]
    polygon = plt.Polygon(vertices, color=color, edgecolor=edge_color, alpha=alpha)
    ax.add_patch(polygon)

# Function to draw genetic material (represented as a helix)
def draw_genetic_material(ax, center, length, spacing, num_turns, color='red'):
    t = np.linspace(0, num_turns * 2 * np.pi, 100)
    x = center[0] + np.cos(t) * spacing
    y = center[1] + np.sin(t) * spacing - length / 2 + t / (num_turns * 2 * np.pi) * length
    ax.plot(x, y, color=color, linewidth=2)

# Main function to simulate the virus structure
def simulate_virus():
    fig, ax = plt.subplots(figsize=(6, 6))
    ax.set_aspect('equal')
    ax.axis('off')

    # Viral envelope (outer circle)
    envelope_center = (0, 0)
    envelope_radius = 2
    draw_circle(ax, envelope_center, envelope_radius)

    # Capsid (icosahedral shape inside the envelope)
    capsid_center = (0, 0)
    capsid_radius = 1.2
    draw_capsid(ax, capsid_center, capsid_radius)

    # Genetic material (helix inside the capsid)
    genetic_center = (0, 0)
    genetic_length = 1.5
    genetic_spacing = 0.2
    genetic_turns = 8
    draw_genetic_material(ax, genetic_center, genetic_length, genetic_spacing, genetic_turns)

    plt.show()

# Run the simulation
simulate_virus()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d.art3d import Poly3DCollection

# Function to draw a sphere (representing the viral envelope)
def draw_sphere(ax, center, radius, color='lightblue', edge_color='blue', alpha=0.5):
    u, v = np.mgrid[0:2*np.pi:20j, 0:np.pi:10j]
    x = center[0] + radius * np.cos(u) * np.sin(v)
    y = center[1] + radius * np.sin(u) * np.sin(v)
    z = center[2] + radius * np.cos(v)
    ax.plot_surface(x, y, z, color=color, edgecolor=edge_color, alpha=alpha)

# Function to draw the capsid as a polyhedron (simple icosahedral representation)
def draw_capsid(ax, center, radius, color='lightgreen', edge_color='green', alpha=0.6):
    phi = (1 + np.sqrt(5)) / 2  # Golden ratio
    vertices = np.array([
        [-1,  phi,  0],
        [ 1,  phi,  0],
        [-1, -phi,  0],
        [ 1, -phi,  0],
        [ 0, -1,  phi],
        [ 0,  1,  phi],
        [ 0, -1, -phi],
        [ 0,  1, -phi],
        [ phi,  0, -1],
        [ phi,  0,  1],
        [-phi,  0, -1],
        [-phi,  0,  1]
    ])
    vertices *= radius / np.linalg.norm(vertices[0])
    vertices += center

    faces = np.array([
        [0, 11, 5], [0, 5, 1], [0, 1, 7], [0, 7, 10], [0, 10, 11],
        [1, 5, 9], [5, 11, 4], [11, 10, 2], [10, 7, 6], [7, 1, 8],
        [3, 9, 4], [3, 4, 2], [3, 2, 6], [3, 6, 8], [3, 8, 9],
        [4, 9, 5], [2, 4, 11], [6, 2, 10], [8, 6, 7], [9, 8, 1]
    ])
    poly3d = [[vertices[vertex_id] for vertex_id in face] for face in faces]
    ax.add_collection3d(Poly3DCollection(poly3d, facecolors=color, edgecolors=edge_color, alpha=alpha))

# Function to draw genetic material (represented as a 3D helix)
def draw_genetic_material(ax, center, length, spacing, num_turns, color='red'):
    t = np.linspace(0, num_turns * 2 * np.pi, 100)
    x = center[0] + np.cos(t) * spacing
    y = center[1] + np.sin(t) * spacing
    z = center[2] + np.linspace(-length / 2, length / 2, 100)
    ax.plot3D(x, y, z, color=color, linewidth=2)

# Main function to simulate the 3D virus structure
def simulate_virus_3d():
    fig = plt.figure(figsize=(8, 8))
    ax = fig.add_subplot(111, projection='3d')
    ax.set_box_aspect([1,1,1])  # Aspect ratio is 1:1:1

    # Viral envelope (outer sphere)
    envelope_center = np.array([0, 0, 0])
    envelope_radius = 2
    draw_sphere(ax, envelope_center, envelope_radius)

    # Capsid (icosahedral shape inside the envelope)
    capsid_center = np.array([0, 0, 0])
    capsid_radius = 1.2
    draw_capsid(ax, capsid_center, capsid_radius)

    # Genetic material (helix inside the capsid)
    genetic_center = np.array([0, 0, 0])
    genetic_length = 2
    genetic_spacing = 0.2
    genetic_turns = 10
    draw_genetic_material(ax, genetic_center, genetic_length, genetic_spacing, genetic_turns)

    # Set view and display
    ax.view_init(elev=20., azim=30)  # Adjust the view angle
    plt.show()

# Run the 3D simulation
simulate_virus_3d()
```
